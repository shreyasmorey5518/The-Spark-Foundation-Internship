### Line plot on the basis of profit made by store

```python
plt.figure(figsize=(10,5))
plt.plot(df['Profit'],color='Green')
plt.grid(alpha=2)
```

### Barplot between Sub-category and category 

```python
# from plot we can say that technology category has highest no sale
# furniture has lowest no. of sale 
plt.figure(figsize=(15,5))
plt.bar('Sub-Category','Category', data=df)
plt.xticks(rotation=90)

plt.show()
```

```python
# Countplot to check highest sale w.r.t various state
plt.figure(figsize=(15,8))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.show()
# we can observe that california has highest no sale w.r.t to other state
# Vyoming  has lowest no. if state
```

```python
# countplot to check highest no. of count of sub-category 
plt.figure(figsize=(12,6))
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
# Blinder has a highest no. of sale as compared to other
```

### Distribution base on Segments and Ship mode

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.countplot(x='Segment',data=df)
plt.title('Distribution of Segments')
plt.subplot(1,2,2)
sns.countplot(x='Ship Mode',data=df)
plt.title('Dstribution of Ship Mode')
plt.show()
```

### Distribution based on Region/Category

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.countplot('Region',data=df)
plt.title('Distribution of Regions')
plt.subplot(1,2,2)
sns.countplot('Region',hue='Category',data=df)
plt.title('Distribution of Categories')
plt.show()
```

### Distribution based on Categories/Sub-Categories

```python
plt.figure(figsize=(20,10))
plt.subplot(2,1,1)
sns.countplot('Category',data=df)
plt.title('Distribution of Categories',fontsize=18)
plt.subplot(2,1,2)
sns.countplot('Sub-Category',data=df)
plt.title('Distribution of Sub-Category',fontsize=18)
plt.show()
```

### Boxplot

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.boxplot(x='Sales', y='Category', data=df, hue='Region')
```
